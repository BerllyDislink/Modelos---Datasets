```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report

```

```python
# Cargar el conjunto de datos
data = pd.read_csv(r"C:\Users\josef\Downloads\DATASETS\Train_data.csv")
```

```python
# Ver los valores únicos de la variable objetivo (class) y sus frecuencias
class_counts = data['class'].value_counts()

# Imprimir los valores únicos y sus frecuencias
print(class_counts)
```

```python
# Separar las características (X) y la variable objetivo (y)
X = data.drop("class", axis=1)
y = data["class"]
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba (70/30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Codificación One-Hot de las características categóricas
X_train_encoded = pd.get_dummies(X_train) # Codifica características de X_train
X_test_encoded = pd.get_dummies(X_test) # Codifica características de X_test
```

```python
# Asegurarse de que ambos conjuntos de datos tienen las mismas características después de codificarlos
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='inner', axis=1)
```

```python
# Crear y entrenar el modelo Naive Bayes
model = GaussianNB()
model.fit(X_train_encoded, y_train)
```

```python
# Hacer predicciones en el conjunto de prueba
y_pred = model.predict(X_test_encoded)
```

```python
# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
```

```python
# Calcular la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Crear y entrenar el modelo GaussianNB
model = GaussianNB()
model.fit(X_train_encoded, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = model.predict(X_test_encoded)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)

# Calcular la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)

# Imprimir la precisión y la matriz de confusión
print("Precisión:", accuracy)
print("Matriz de confusión:\n", conf_matrix)

# Calcular e imprimir el reporte de clasificación
report = classification_report(y_test, y_pred)
print("Reporte de clasificación:\n", report)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import OneHotEncoder
```

```python

# Cargar el conjunto de datos
data = pd.read_csv(r"C:\Users\josef\Downloads\DATASETS\Train_data.csv")

# Separar las características (X) y la variable objetivo (y)
X = data.drop("class", axis=1)
y = data["class"]
```

```python


# Dividir los datos en conjuntos de entrenamiento y prueba (70/30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python


# Crear una instancia de OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python


# Ajustar y transformar X_train y X_test
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)
```

```python


# Crear y entrenar el modelo MultinomialNB
multinomial_nb_model = MultinomialNB()
multinomial_nb_model.fit(X_train_encoded, y_train)
```

```python
# Hacer predicciones con el modelo MultinomialNB
y_pred_multinomial = multinomial_nb_model.predict(X_test_encoded)

# Evaluar el modelo MultinomialNB
accuracy_multinomial = accuracy_score(y_test, y_pred_multinomial)
conf_matrix_multinomial = confusion_matrix(y_test, y_pred_multinomial)
report_multinomial = classification_report(y_test, y_pred_multinomial)
```

```python
print("\nModelo MultinomialNB")
print("Precisión:", accuracy_multinomial)
print("Matriz de confusión:\n", conf_matrix_multinomial)
print("Reporte de clasificación:\n", report_multinomial)
```
